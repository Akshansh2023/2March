Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

# Answer 
Matplotlib is a popular Python library used for creating high-quality 2D and 3D visualizations of data. It provides a wide range of functions for generating different types of plots and charts, making it a versatile tool for data visualization.

Matplotlib is used for a variety of purposes, such as:

Exploratory data analysis
Communicating research findings
Presenting data in a clear and concise manner
Teaching data visualization and analysis
Creating visualizations for reports and presentations
The Pyplot module of Matplotlib is a collection of functions that provides a convenient interface for creating plots and charts. Here are five types of plots that can be created using Pyplot:

Line plot: This type of plot is used to visualize trends and patterns in data over time.
Scatter plot: This type of plot is used to show the relationship between two variables.
Histogram: This type of plot is used to visualize the distribution of a dataset.
Bar chart: This type of plot is used to compare values across different categories.
Pie chart: This type of plot is used to show the proportion of different categories in a dataset.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

# Answer

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.xlabel(" X direction " )
plt.ylabel(" Y direction ")
plt.title(" Random Ploting ")
plt.scatter(x,y)
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

# Answer 3

The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows us to display several plots in a grid-like pattern, making it easier to compare and analyze different aspects of our data.

Here's an example of how to use the subplot() function to create four line plots in a 2x2 grid:


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig,axs = plt.subplots(nrows=2, ncols=2)

# Plot the first subplot
axs[0, 0].plot(x1, y1, color='blue')
axs[0, 0].set_title('Line 1')

# Plot the second subplot
axs[0, 1].plot(x2, y2, color='red')
axs[0, 1].set_title('Line 2')

# Plot the third subplot
axs[1, 0].plot(x3, y3, color='green')
axs[1, 0].set_title('Line 3')

# Plot the fourth subplot
axs[1, 1].plot(x4, y4, color='purple')
axs[1, 1].set_title('Line 4')

# Add a main title for the entire figure
fig.suptitle('Four Line Plots')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.3, hspace=0.5)

# Show the plot
plt.show()
```

import numpy as np

# Answer

A bar plot is a chart that represents categorical data with rectangular bars, where the height or length of the bar corresponds to the value of the category it represents. Bar plots are commonly used to compare the values of different categories, and are especially useful when the number of categories is relatively small.
The matplotlib.pyplot module provides a bar() function to create bar plots. The bar() function takes two arguments: the x-coordinates of the bars (in this case, the company names), and the y-coordinates of the bars (in this case, the profits).
Here's an example of how to create a vertical bar plot using the given data:

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company , profit)
```

# question 5
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Answer 

A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the median, quartiles, and outliers of the data. 
The box represents the interquartile range (IQR), which is the middle 50% of the data, and the whiskers extend to the minimum and maximum values that are not outliers.
Box plots are used to display the distribution and variability of a dataset, and are particularly useful when comparing multiple datasets.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.xticks([1, 2], ["Box 1", "Box 2"])
plt.show()
```
